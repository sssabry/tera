```python
from Benchmarks import FLOWSTAR_benchmarks
from TERA.Workbench.Task import TaskRunner
from TERA.Workbench.Visualizer import Visualizer
from TERA.Workbench.Report import Report

tasks = FLOWSTAR_benchmarks.get_chen_hybrid()
```

### Bouncing Ball (Chen - Figure 4.15)

```python
task = tasks["hybrid_bouncing_ball"]

result = TaskRunner.run(task)

plot = Visualizer.plot(result, dims=(0, 1), mode="phase", title="Bouncing Ball: Phase Portrait", filled=False)
plot = Visualizer.plot(result, dims=(0,1), mode="time", y_idx=0, title="Bouncing Ball: Time Evolution", filled=False)
```

```python
mode_colors = {'up': 'magenta', 'down': 'blue'}
plot = Visualizer.plot(result, dims=(0,1), mode="time", mode_colors=mode_colors, y_idx=0, title="Bouncing Ball: x over Time", filled=True, legend_fontsize=16)
```
