```python
from Benchmarks import C2E2_benchmarks
from TERA.Workbench.Task import TaskRunner
from TERA.Workbench.Visualizer import Visualizer

tasks = C2E2_benchmarks.get_c2e2_continuous()
```

## C2E2: Examples

### Jet Engine

Exact match with step size = 0.01 & order = 10

```python
result_jet= TaskRunner.run(tasks["cont_jetengine"], progress_bar=True)
plot = Visualizer.plot(result_jet, title="Jet Engine: x vs y")
```

Minor modification: Step size = 0.05

```python
jetengine = tasks["cont_jetengine"]
jetengine.step_size = 0.05

result_jet= TaskRunner.run(jetengine, progress_bar=True)
plot = Visualizer.plot(result_jet, title="Jet Engine: x vs y")
```

Left Right Mode

```python
jetengine = tasks["cont_jetengine"]
jetengine.engine_params.update({"setting": "left_right"})

result_jet= TaskRunner.run(jetengine, progress_bar=True)
plot = Visualizer.plot(result_jet, title="Jet Engine: x vs y")
```
