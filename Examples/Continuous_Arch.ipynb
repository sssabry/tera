# 2017 Nonlinear Continuous

### Van der Pol Oscillator

```sage
task_vdp = arch_2017["arch17_cont_vanderpol"]
result_vdp = TaskRunner.run(task_vdp)
plot = Visualizer.plot(result_vdp, title="Van der Pol (ARCH 17)")
```

### Laub-Loomis Model

Laub Loomis: Order = 3, Step size = 0.07

```sage
task_ll = arch_2017["arch17_cont_laubloomis"]
task_ll.step_size = 0.07
task_ll.order=3
result_ll = TaskRunner.run(task_ll)
plot = Visualizer.plot(result_ll, dims=(0, 2), title="Laub Loomis: x1 vs x3")
plot = Visualizer.plot(result_ll, dims=(0, 2), title="Laub Loomis: x1 vs x3", edgecolor="blue")
timeplot = Visualizer.plot(result_ll, mode="time", y_idx=3, title="Laub Loomis: x_4 over Time")
```

Laub Loomis: Order = 4, Step size = 0.05

```sage
task_ll = arch_2017["arch17_cont_laubloomis"]
result_ll = TaskRunner.run(task_ll)
plot = Visualizer.plot(result_ll, dims=(0, 2), title="Laub Loomis: x1 vs x3")
timeplot = Visualizer.plot(result_ll, mode="time", y_idx=3, title="Laub Loomis: x_4 over Time")
```
