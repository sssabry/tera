```sage
from Benchmarks import C2E2_benchmarks
from TERA.Workbench.Task import TaskRunner
from TERA.Workbench.Visualizer import Visualizer

tasks = C2E2_benchmarks.get_c2e2_hybrid()
```

## C2E2: Examples

### Cardiac Cell

```sage
task = tasks["hybrid_cardiac"]

result = TaskRunner.run(task, progress_bar=True)

mode_colors = {'stimOn': 'magenta', 'stimOff': 'blue'}
plot = Visualizer.plot(result, dims=(0,1), mode="time", mode_colors=mode_colors, y_idx=0, title="Cardiac Cell: u over Time", filled=True)
```

### Linear Thermometer

```sage
task = tasks["hybrid_linthermo_v1"]
result = TaskRunner.run(task, progress_bar=True)

mode_colors = {'heating': 'red', 'cooling': 'blue'}
plot = Visualizer.plot(result, mode="time", mode_colors=mode_colors, y_idx=0, title="Thermostat: x over Time", filled=True)
```
